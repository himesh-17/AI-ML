```python
import numpy as np
import time
```

```python
#comparison between py list and np arrays
size = 10_000_000

#py list 
start = time.time()
py_list = list(range(size))
sq = [x**2 for x in py_list]
end = time.time()

print(f"Python list time {end -start} second")

#np array

start = time.time()
np_arr = np.arange(size)

#vectorization
sq_arr = np_arr ** 2
end = time.time()
print(f"NumPY Array time {end -start} second")
```

```python
#memory 
import sys
print(f"py list size {sys.getsizeof(py_list) * len(py_list)} bytes")
print(f"np array size {sys.getsizeof(np_arr.nbytes)} bytes")
```
