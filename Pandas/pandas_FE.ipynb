```python
import pandas as pd 
```

```python
df = pd.read_csv("raw_data.csv")
```

```python
# Trnsformation / Feature Engineering

df2 = df.copy()

# apply a fnx to a row or col
df2["tax"] = df["income"].apply(lambda x :  (x* 20 )/ 100 if x >= 50000 else (x * 10 )/ 100  ) 

# map values to another
gender_map = { "Male" : "M" , "Female" : "F" , "Unknown" : "U"}
df2["gender"] = df2["gender"].map(gender_map)

# add a new col and  assign some new values
df2 = df2.assign(new_income = df2["income"] * 1.1) #10% increment in salary

# replace (old , new) old vale to new value 
df2["country"] = df2["country"].replace("USA" , "US")
df2
```

```python
# Transformation - Sorting 

df2  = df.copy();

#renaming 
df2.columns = ["Id", "Name", "Age", "Income", "Gender", "Country", ]
df2 = df2.rename(columns={"Income" : "Salary"})
df2 =df2.rename(index = {0 :"first"})

# ## Sorting - values & index
df2.sort_values("Income")                     # sort values in ascending
df2.sort_values("Income", ascending=False)    # sort values in descending
df2.sort_values(["Age", "Income"])            # sorts for age, if age is same then sorts for income 

sorted_df2 = df2.sort_values(["Age", "Income"])
sorted_df2.sort_index()

# Reset Index
sorted_df2.reset_index()
sorted_df2.reset_index(drop=True)     # to drop original index vals

# # Ranking
df2["Ranking"] = df2["Income"].rank(ascending=False, method="dense")

# # Reorder
df2[["Id", "Name", "Age", "Income", "Gender", "Country", "Ranking"]]


df2
```

```python
# Writing to CSV files 
import numpy as np

df = pd.read_csv("raw_data.csv")


df2= df.copy()
df2 = df2.drop_duplicates()

df2 = df2.fillna(0)

df2["country"] = df2["country"].replace(0, np.nan).ffill() 
df2["age"]= df2["age"].replace(0,df2["age"].mean())

df2 = df2.sort_values("age")

df2 =df2.reset_index(drop = True)

df2.to_csv("sorted_data.csv")

```
