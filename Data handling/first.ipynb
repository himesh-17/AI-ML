```python
import json
#Load data
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)

    return data

data = load_data("data.json")
```

```python
# cleaning 
def clean_data(data):
    for user in data :
        if(type(user["age"]) == str ):
            user["age"] = int(user["age"])
        elif(user["age"] ==None):
            user["age"] = 0
        

        if(user["rating"] == str):
            user["rating"] = float(user["rating"])

        print(type(user["age"]) ,type(user["rating"]))



        
```
